```python
# Define the function
def classify(values, boundary=0):
    """ 
    Classifies values as being: 
    larger (True) or equal/smaller (False) than a boundary.
    """
    return [(True if v > boundary else False) for v in values]
```

```python
import numpy as np

my_values = np.random.uniform(0.0, 1.0,1000)

# Call it with a different boundary value
classify(my_values, boundary=0.5)
```
